```python
from PIL import Image
import numpy as np
import pandas as pd
import os

# load images into pandas dataframe

def load_images(path, width, height):
    dataset = []
    labels = []
    
    for classname in os.listdir(path):
        subdir = os.path.join(path, classname)
        
        for img_name in os.listdir(subdir):
            # read image and convert to 1d array
            img_path = os.path.join(subdir, img_name)
            img = np.asarray(Image.open(img_path).resize((width, height))).flatten()
            
            dataset.append(img)
            labels.append(classname)
            
    df = pd.DataFrame(data = np.array(dataset))
    
    df['label'] = labels
    
    return df
```

```python
path = r"E:\hand_gesture_dataset\whole"

df = load_images(path, 150, 150)
```
