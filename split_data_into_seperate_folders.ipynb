create train, test, validation folders

```python
import os, shutil

# assuming this directories exist
original_dataset_dir = r'E:\hand_gesture_dataset'
base_dir = r'E:\hand_gesture_dataset\splited'

# create train, validation, test datasets
# asuming the directories don't exist
train_dir = os.path.join(base_dir, 'train')
os.mkdir(train_dir)

validation_dir = os.path.join(base_dir, 'validation')
os.mkdir(validation_dir)

test_dir = os.path.join(base_dir, 'test')
os.mkdir(test_dir)
```

move images from original directories to the train directory, create subdirectories with the correponding category name

```python
for dir in os.listdir(original_dataset_dir):
    if not dir.startswith("handgesturedataset"):
        continue
    
    path = os.path.join(original_dataset_dir, dir)
    
    for filename in os.listdir(path):
        # the category name is the 6th character in the inage name
        class_dir_name = filename[6]
        class_dir_path = os.path.join(train_dir, class_dir_name)
        
        if not os.path.exists(class_dir_path):
            os.mkdir(class_dir_path)
            os.mkdir(os.path.join(test_dir, class_dir_name))
            os.mkdir(os.path.join(validation_dir, class_dir_name))
            
        # calculating the number of files in source dir (not a very good way, think a better silution (use caching instead))
        source_file_name_ext = len([name for name in os.listdir(class_dir_path)])
        
        # copy for handgesturedataset_* to train directory
        src = os.path.join(path, filename)
        dst = os.path.join(class_dir_path, class_dir_name + '.' + str(source_file_name_ext) + '.png')
        shutil.copyfile(src, dst)   
        
```

train, test, validation split 

```python
import random

# choose random 10 for test and random 10 for validation for each class
test_size_for_each_class = 14
validation_size_for_each_class = 10

for classname in os.listdir(train_dir):
    class_dir_path = os.path.join(train_dir, classname)
    val_class_dir_path = os.path.join(validation_dir, classname)
    test_class_dir_path = os.path.join(test_dir, classname)
    
    # some logs might be useful here. like "moving from 'src' to 'dst'""
    print(class_dir_path)
    print(val_class_dir_path)
    print(test_class_dir_path)
    
    # move to test directory
    for i in range(test_size_for_each_class):
        src_file_name = random.choice(os.listdir(class_dir_path))
        src = os.path.join(class_dir_path, src_file_name)
        dst = test_class_dir_path
        shutil.move(src, dst)   

    # move to validation directory
    for i in range(validation_size_for_each_class):
        src_file_name = random.choice(os.listdir(class_dir_path))
        src = os.path.join(class_dir_path, src_file_name)
        dst = val_class_dir_path
        shutil.move(src, dst)
    
```
